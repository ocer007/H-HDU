```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from random import randint

# txt -> pandas
musae_ENGB = pd.read_csv(
    "/content/sample_data/musae_ENGB_edges.csv.gz",
    compression="gzip",
    sep=",",
    names=["start", "end"],
)

# pandas->network
G = nx.from_pandas_edgelist(musae_ENGB, "start", "end")


# 1 Average path length
# 平均路径长度 Average path length
def average_path_length():
    result = nx.average_shortest_path_length(G)
    return result


# 最短路径分布可视化
def show_path_length_table():
    # 最短路径字典
    shortest_path_lengths = dict(nx.all_pairs_shortest_path_length(G))
    # 直径
    diameter = max(nx.eccentricity(G, sp=shortest_path_lengths).values())

    x = list(range(diameter + 1))
    y = []
    for i in range(diameter + 1):
        y.append(0)
    for pls in shortest_path_lengths.values():
        for plss in pls.values():
            y[plss] += 1
    yy = [i * 100 / sum(y) for i in y]
    plt.figure(figsize=(15, 8))
    plt.bar(x,y)
    plt.title("distribution of shortest path length", fontdict={"size": 35}, loc="center")
    plt.xlabel("Shortest Path Length", fontdict={"size": 22})
    plt.ylabel("Frequency (%)", fontdict={"size": 22})
    plt.show()
    return


# 2 Degree distribution
# 平均度
def average_degree():
    dict_degree = dict(nx.degree(G))  # 度分布字典
    result = sum(dict_degree.values()) / len(dict_degree)
    return result


# 度分布可视化
def show_degree_distribution():
    hist_degree = nx.degree_histogram(G)  # 度分布频数表
    dict_degree = dict(nx.degree(G))
    x = list(range(max(dict_degree.values()) + 1))
    y = [i * 100 / sum(hist_degree) for i in hist_degree]
    plt.figure(figsize=(10, 6))
    plt.bar(x, y)
    plt.xlabel("Degree", fontdict={"size": 22})
    plt.ylabel("Frequency (%)", fontdict={"size": 22})
    plt.xlim([0, max(dict_degree.values()) + 1])
    plt.show()
    return


# 3 clustering coefficient
# 平均聚类系数
def average_clustering_G():
    result = nx.average_clustering(G)
    return result


# 聚类系数分布
def show_clustering_distribution():
    plt.figure(figsize=(10, 6))
    plt.hist(nx.clustering(G).values(), bins=25)
    plt.title("clustering_distribution", fontdict={"size": 35}, loc="center")
    plt.xlabel("clustering coefficient", fontdict={"size": 20})
    plt.ylabel("number", fontdict={"size": 20})
    plt.show()
    return


# 4 Assortativity
# 同配系数
def assortativity_G():
    result = nx.degree_assortativity_coefficient(G)
    return result



# 5 Network community
# LPA标签传播社区发现算法
def LPA_for_network_community():
    label_propagation_communities = nx.community.label_propagation_communities(G)
    counter = len(label_propagation_communities)
    colors_for_nodes = []
    for i in range(G.number_of_nodes()):
        colors_for_nodes.append("")

    for com in label_propagation_communities:
        color = '#%06X' % randint(0, 0xFFFFFF)  # creates random RGB color
        for node in list(com):
            colors_for_nodes[node] = color

    pos = nx.spring_layout(G, iterations=6)
    plt.figure(figsize=(10, 6))
    title = "%d communities found in total with LPA" % counter
    plt.title(title, fontdict={"size": 20}, loc="center")
    nx.draw_networkx(G, pos=pos, node_size=1, with_labels=False, width=0.01, node_color=colors_for_nodes)
    plt.show()
    return


# greedy_modularity社区算法
def greedy_modularity_for_network_community():
    greedy_modularity_communities = nx.community.greedy_modularity_communities(G)
    counter = len(greedy_modularity_communities)
    colors_for_nodes = []
    for i in range(G.number_of_nodes()):
        colors_for_nodes.append("")
    for com in greedy_modularity_communities:
        color = '#%06X' % randint(0, 0xFFFFFF)
        for node in list(com):
            colors_for_nodes[node] = color

    pos = nx.spring_layout(G, iterations=6)
    plt.figure(figsize=(10, 6))
    title = "%d communities found in total with greedy_modularity" % counter
    plt.title(title, fontdict={"size": 20}, loc="center")
    nx.draw_networkx(G, pos=pos, node_size=10, with_labels=False, width=0.15, node_color=colors_for_nodes)
    plt.show()
    return


# Louvain Fast community unfolding算法
def fast_community_unfolding_for_network_community():
    fast_community_unfolding_communities = nx.community.louvain_communities(G)
    counter = len(fast_community_unfolding_communities)
    colors = []
    for i in range(G.number_of_nodes()):
        colors.append("")
    for com in fast_community_unfolding_communities:
        color = '#%06X' % randint(0, 0xFFFFFF)
        for node in list(com):
            colors[node] = color

    pos = nx.spring_layout(G, iterations=6)
    plt.figure(figsize=(10, 6))
    title = "%d communities found in total with Louvain Fast community unfolding" % counter
    plt.title(title, fontdict={"size": 20}, loc="center")
    nx.draw_networkx(G, pos=pos, node_size=10, with_labels=False, width=0.15, node_color=colors)
    plt.show()
    return


# asyn_fluidc算法
def asyn_fluidc_for_network_community(counter):
    asyn_fluidc_communities = nx.community.asyn_fluidc(G, counter)
    colors = []
    for i in range(G.number_of_nodes()):
        colors.append("")
    for com in asyn_fluidc_communities:
        color = "#%06X" % randint(0, 0xFFFFFF)  # creates random RGB color
        for node in list(com):
            colors[node] = color

    pos = nx.spring_layout(G, iterations=6)
    plt.figure(figsize=(10, 6))
    title = "%d communities found in total with Louvain Fast community unfolding" % counter
    plt.title(title, fontdict={"size": 20}, loc="center")
    nx.draw_networkx(G, pos=pos, node_size=10, with_labels=False, width=0.15, node_color=colors)
    plt.show()
    return


# 图的可视化
def show_graph():
    pos = nx.spring_layout(G, iterations=6)
    plt.figure(figsize=(10, 6))
    title = "Visualize the whole graph connectivity"
    plt.title(title, fontdict={"size": 20}, loc="center")
    nx.draw_networkx(G, pos=pos, node_size=1, with_labels=False, width=0.02)
    plt.show()
    return

# 1  Average path length
# 平均路径长度
print("average_path_length:",average_path_length())
# 最短路径分布可视化
show_path_length_table()
# 2 Degree distribution
# 平均度
print("average_degree:",average_degree())
# 度分布可视化
show_degree_distribution()

# 3 Clustering Coefficient
# 平均聚类系数
print("average_clustering:",average_clustering_G())
# 聚类系数可视化
show_clustering_distribution()
# 4 Assortativity
# 同配系数
print("assortativity: ",assortativity_G())

# 5 Network community
# LPA算法 标签传播社区发现算法
LPA_for_network_community()
# greedy_modularity社区算法
greedy_modularity_for_network_community()
# Louvain Fast community unfolding算法
# fast_community_unfolding_for_network_community()
# asyn_fluidc算法
asyn_fluidc_for_network_community(10)
# 6 Visualize the whole graph connectivity.
# 连通分量的个数
print(nx.number_connected_components(G))
# 图的可视化
show_graph()
```
